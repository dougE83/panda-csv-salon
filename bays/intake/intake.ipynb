## Pandas 3.x — `read_csv()` String Dtype

In pandas 3.x, string columns are now `StringDtype` (Arrow-backed) by default, not `object`.

This block shows how to check for string columns the right way.

```python
# Detect string columns — works correctly in pandas 3.x
string_cols = [col for col in df.columns if pd.api.types.is_string_dtype(df[col])]
print("String columns:", string_cols)

# Show their actual dtype
for col in string_cols:
    print(f"  {col}: {df[col].dtype}")
```

## Pandas 3.x — Copy-on-Write

Copy-on-Write is on by default in pandas 3.x. This block demonstrates the correct assignment pattern.

```python
# CORRECT: always reassign when modifying
# df = df[df['some_column'] > 0]  # reassignment, safe

# INCORRECT (old pandas pattern — silently broken in 3.x):
# df[df['some_column'] > 0]['other_column'] = 'value'  # chained assignment

# Use .loc for conditional updates:
# df.loc[df['some_column'] > 0, 'other_column'] = 'value'

print("Copy-on-Write is ON by default in pandas 3.x.")
print(f"pd.options.mode.copy_on_write = {pd.options.mode.copy_on_write}")
```

## Script Generator

Delete any blocks above that you don't want in the output, then run this cell.
A clean `.py` script will be saved to the `scripts/` folder.

```python
# ============================================================
# SCRIPT GENERATOR
# ============================================================
import nbformat
from pathlib import Path

NOTEBOOK_NAME = "intake.ipynb"

nb = nbformat.read(NOTEBOOK_NAME, as_version=4)

code_cells = [
    cell for cell in nb.cells
    if cell.cell_type == "code"
    and cell.source.strip()
    and "# SCRIPT GENERATOR" not in cell.source
]

script = "\n\n".join(cell.source for cell in code_cells)

output_dir = Path("../../scripts")
output_dir.mkdir(exist_ok=True)
output_file = output_dir / f"{Path(NOTEBOOK_NAME).stem}_script.py"
output_file.write_text(script, encoding="utf-8")

print(f"Script saved to: {output_file.resolve()}")
```
