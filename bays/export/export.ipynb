# Export Bay — Write Your Data Out

Save your processed DataFrame to a file. Pick the format that fits where this data is going next.

See the [Export README](README.md) for help choosing a format.

```python
# ============================================================
# SALON CONFIG — edit the line below to point to your CSV file
# ============================================================
import pandas as pd
from pathlib import Path

PARKING_BAY = Path("../../parking_bay")
CSV_FILE = "your_file.csv"  # <-- change this to your input file name

df = pd.read_csv(PARKING_BAY / CSV_FILE)
print(f"Loaded: {CSV_FILE}  |  Shape: {df.shape}")
```

## Export to CSV

The most universal format. Works everywhere.

```python
OUTPUT_FILE = PARKING_BAY / "output.csv"  # <-- change the output file name

df.to_csv(OUTPUT_FILE, index=False)
print(f"Saved: {OUTPUT_FILE.resolve()}")
print(f"Rows: {len(df)} | Columns: {len(df.columns)}")
```

## Export to Excel

Best for sharing with non-technical stakeholders. Requires `openpyxl` (included in this repo).

```python
OUTPUT_FILE = PARKING_BAY / "output.xlsx"  # <-- change the output file name

df.to_excel(OUTPUT_FILE, index=False, sheet_name="Data")
print(f"Saved: {OUTPUT_FILE.resolve()}")

# To write multiple sheets:
# with pd.ExcelWriter(OUTPUT_FILE, engine="openpyxl") as writer:
#     df.to_excel(writer, sheet_name="Data", index=False)
#     summary_df.to_excel(writer, sheet_name="Summary", index=False)
```

## Export to JSON

Best for web applications or passing data to an API.

```python
OUTPUT_FILE = PARKING_BAY / "output.json"  # <-- change the output file name

# orient='records' produces [{col: val, ...}, ...] — most common format
df.to_json(OUTPUT_FILE, orient="records", indent=2)
print(f"Saved: {OUTPUT_FILE.resolve()}")

# Other orientations:
# orient='split'   — {index: [], columns: [], data: []}
# orient='index'   — {row_index: {col: val, ...}, ...}
# orient='columns' — {col: {row_index: val, ...}, ...}
```

## Export to Parquet

Best for large datasets or downstream data pipelines. Preserves dtypes exactly, including pandas 3.x Arrow-backed string types. Much smaller file size than CSV for large data.

```python
OUTPUT_FILE = PARKING_BAY / "output.parquet"  # <-- change the output file name

df.to_parquet(OUTPUT_FILE, index=False)
print(f"Saved: {OUTPUT_FILE.resolve()}")

# Verify the round-trip:
# df_check = pd.read_parquet(OUTPUT_FILE)
# print(f"Re-loaded shape: {df_check.shape}")
# print(df_check.dtypes)
```

## Script Generator

Delete any blocks above that you don't want in the output, then run this cell.
A clean `.py` script will be saved to the `scripts/` folder.

```python
# ============================================================
# SCRIPT GENERATOR
# ============================================================
import nbformat
from pathlib import Path

NOTEBOOK_NAME = "export.ipynb"

nb = nbformat.read(NOTEBOOK_NAME, as_version=4)

code_cells = [
    cell for cell in nb.cells
    if cell.cell_type == "code"
    and cell.source.strip()
    and "# SCRIPT GENERATOR" not in cell.source
]

script = "\n\n".join(cell.source for cell in code_cells)

output_dir = Path("../../scripts")
output_dir.mkdir(exist_ok=True)
output_file = output_dir / f"{Path(NOTEBOOK_NAME).stem}_script.py"
output_file.write_text(script, encoding="utf-8")

print(f"Script saved to: {output_file.resolve()}")
```
